```python
from nltk.tokenize import  word_tokenize
```

```python
data["OriginalTweetTokenized"]=data["OriginalTweet"].apply(lambda x: word_tokenize(x))
```

```python
def lexical_diversity(text):
    return len(text)/len(set(text))
data["LexicalDiversity"]=data["OriginalTweetTokenized"].apply(lambda x: lexical_diversity(x))
```
